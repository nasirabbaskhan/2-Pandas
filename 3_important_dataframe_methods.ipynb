##### apply()

```python
# culculate the sigmoid of all numbers

def sigmoid(value):
    return 1/(1 + np.exp(-value))

temp.apply(sigmoid)
```

```python
# apply on dataframe

points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2st point':[(3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
# apply on rows in dataframe  -> logic
# custom logic
def euclidean(row):
    point_A= row['1st point']
    point_B= row['2st point']
    
    return ((point_A[0] - point_B[0])**2 + (point_A[1] - point_B[1])**2)**0.5
    
```

```python
# axis=1  -> rows
# axis=0  -> colmn
euclidean_distance = points_df.apply(euclidean, axis=1) 
euclidean_distance
```

```python
# add Distance col in df
points_df["Distance"]= euclidean_distance
```
