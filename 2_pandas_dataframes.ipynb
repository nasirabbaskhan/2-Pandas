```python
# toss winner is match winner in percentage

win_with_toss= ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]


lose_with_toss = ipl[ipl['TossWinner']!=ipl['WinningTeam']].shape[0]


total_maches = ipl.shape[0]
# lose_with_toss= total_maches-win_with_toss

win_with_toss_percentage = (100*win_with_toss)/total_maches
loose_with_toss_percentage = (100*lose_with_toss)/total_maches

print("total_maches: ", total_maches)
print("win_with_toss: ", win_with_toss)
print("lose_with_toss: ", lose_with_toss)
print("win_with_toss_percentage: ", win_with_toss_percentage)
print("loose_with_toss_percentage: ", loose_with_toss_percentage)


# ipl.head(2)
```

```python
# write a fuction that can return the track record of 2 teams against each others
def track_record(Team1:str,Team2:str):
    first = (ipl['Team1']==Team1) & (ipl['Team2']==Team2)
    second =(ipl['Team2']==Team1) & (ipl['Team1']==Team2)
    all_maches = ipl[first | second]
    return all_maches
    
result =track_record("Royal Challengers Bangalore","Rajasthan Royals")
result
# ipl
```
