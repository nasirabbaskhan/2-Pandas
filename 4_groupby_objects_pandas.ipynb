#### GroupBy Attributes and Methodes

```python
# len  -> find total numbers of group
# size -> find items in each group (catagory)
# first() / last()
# get_group -> vs filtering
# group
# describe
# sample
# nunique
```

##### len
find total numbers of group

```python
len(movies.groupby("Genre"))  # retuen total number of groups
```

##### size
find items in each group (catagory)

```python
movies.groupby("Genre").size() # retuen total number of rows in each groups
```

```python
movies['Genre'].value_counts()  # as same
```

##### first() / last() / nth()


```python
movies.groupby("Genre").first() # return first row of each groups 
```

```python
movies.groupby("Genre").last() # return last row of each groups 
```

```python
movies.groupby("Genre").nth(5) # retun your provided row of each group
```

##### get_group vs filtering

```python
movies["Genre"] == "Comedy" # filtering with logic
movies.groupby("Genre").get_group("Comedy") # return all the rows of your provided group (Comedy)
```

##### groups
it return all indexes of each groups

```python
movies.groupby("Genre").groups
```

##### describe / sample / nunique

```python
movies.groupby("Genre").describe() # return the statistical analysis of each group
```

```python
movies.groupby("Genre").sample()  # return 1 reandomly row of each group
```

```python
movies.groupby("Genre").nunique()  # return the number of unique rows
```

#### agg method
used to apply multiples aggregation on each cols of each groups

```python
generes = movies.groupby("Genre")

generes.sum() # apply single aggregation 
```

```python
generes = movies.groupby("Genre")  # applying aggregation on each numarical colum (multiple aggregation) 
generes.agg(
    {
        'Runtime':"mean",
        'IMDB_Rating':"mean",
        'No_of_Votes':"sum",
        'Gross':"sum",
        'Metascore':"min",
    
    }
)
```

```python
# apply multiples aggregation on each numarical column
# passing list

generes.agg(['min','max','sum'])

```

```python
# adding both the syntex
generes.agg(
    {
        'Runtime':['min',"mean"],
        'IMDB_Rating':['min',"mean"],
        'No_of_Votes':['min',"sum"],
        'Gross':['min',"sum"],
        'Metascore':['min',"mean"],
    
    }
)
```

#### split (apply) combine

```python
generes.apply(min)
```

```python
# find number of movies starting with A for each group  ->logic
generes = movies.groupby("Genre") # splitting
# generes["Series_Title"].str.startswith("A")  -> it only apply on one group not on all groups

def foo(group):
    group =group[group["Series_Title"].str.startswith("A")].nunique()
    return group
```

```python
generes.apply(foo)
```

```python
# find the ranking of each movies in the group acordig to the IMDB_Rating  ->logic
def ranking(group):
    # group["Metascore"].rank()
    group["genre_rank"]= group["IMDB_Rating"].rank(ascending=True)
    # print(group)
    return group
    
```

```python
generes.apply(ranking)
```

```python
# find the normalized rating on the bases of IMDB_Rating   ->logic
def normal(group):
    group["normal_rank"] = (group["IMDB_Rating"]-group["IMDB_Rating"].min())/(group["IMDB_Rating"].max()-group["IMDB_Rating"].min())
    return group
```

```python
generes.apply(normal)
```

```python
# find the highest rated movies of each genre -> logic


def rated(group):
    group = group[group['IMDB_Rating']== group['IMDB_Rating'].max()] 
    return group
```

```python
generes.apply(rated)
```

```python
# find top ten batsman interm of runs ->logic
ipl.groupby('batsman')["batsman_runs"].sum().sort_values(ascending=False).head(10)
```

```python
# find  batsman with max number of 6  ->logic

new_df =ipl[ipl["batsman_runs"]==6] # find df with only 6 

new_df.groupby("batsman")['batsman_runs'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs  ->logic

new_df = ipl[(ipl['over'] >15) & ((ipl['batsman_runs']== 6) | (ipl['batsman_runs'] == 4 ))]
new_df.groupby("batsman")["batsman_runs"].count().sort_values(ascending=False).head(1)
```

```python
# find V kohly's record againt all the team ->logic

temp_df = ipl[ipl["batsman"]=="V Kohli"]
temp_df.groupby("bowling_team")["batsman_runs"].sum().reset_index().sort_values("batsman_runs", ascending=False)
```

```python
# # create a function that can return the highest score of any batsman ->logic
temp_df= ipl[ipl["batsman"]=='CH Gayle']
temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def h_score(batsmain):
    temp_df= ipl[ipl["batsman"]==batsmain]
    return temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
    
```

```python
h_score("MS Dhoni")
```

```python
# create a function that can return the highest score of each batsman ->logic
def highest_score(group):
    match_id = group.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1)
    return match_id

# ipl
```

```python
ipl.groupby("batsman").apply(highest_score).sort_values(ascending=False)
```
