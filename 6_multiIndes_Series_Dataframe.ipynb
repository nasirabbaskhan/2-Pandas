##### MultiIndex Series
it is also known as Hierarchical Indexing

Multiple index levels withen a single Index

```python
# how to create multiindexing object
# pd.MultiIndex.from_tuples()
# pd.MultiIndex.from_product() ->easy meathod

index_values = [('sce',2019),('sce',2020),('sce',2021),('sce',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
pd.MultiIndex.from_tuples(index_values)

```

```python
multiindex = pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]]) # second method
```

```python
# create a Series with Multiindexing object
series = pd.Series([1,2,3,4,5,6,7,8], index=multiindex)
series
```

```python
# how to fech data
series.loc[('cse',2020)] # -> here multiindex seris is 2D
```

```python
# convert the multiindex seris in Dataframe
#unstack()

df = series.unstack()
df
```

```python
# convert any Dataframe into multiIndex series

df.stack()
```

#### Wide vs Long Data

**Wide format**: is where whe have a single rows for every data point with multiple columns to hold the values of various attribut.

**Long format**: is where, for each datapoint whe have many rows as the number of attributes and each row contains the value of a particular attributes for a given data.



##### Melt
convert the wide Data into Long Data

```python
## Melt   -> wide to long

temp = pd.DataFrame({'cse':[120]})
temp
```

```python
temp.melt()
```

```python
temp1 = pd.DataFrame({'cse':[120],'ece':[100],'mech':50})
temp1
```

```python
temp1.melt()
```

```python
temp1.melt(var_name='Brances',value_name="students")
```

```python
temp3 = pd.DataFrame(
    {
        'branch':['cse','ece','mech'],
        '2020':[100,150,60],
        '2021':[120,130,80 ],
        '2022':[150,1140,70]
    }
)
temp3
```
